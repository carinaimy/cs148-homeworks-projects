```python
def plot_decision_boundaries(X, y, model_class, **model_params):
    """
    Resource from:
    https://gist.github.com/anandology/772d44d291a9daa198d4
    http://scikit-learn.org/stable/auto_examples/ensemble/plot_voting_decision_regions.html
    http://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_digits.html
    """
    reduced_data = X[:, :2]
    model = model_class(**model_params)
    model.fit(reduced_data, y)
    h = .02 # Mesh Step size of the mesh. 

    # Plot the decision boundary.
    x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
    y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Obtain labels for each point in mesh using the model.
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])    

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8, cmap=ListedColormap(['#FF0000', '#0000FF']))
    return plt
```

```python
X = np.array([[0, 2], [1, 4], [3, 6], [3, 2], [4, 4], [2, 0]])
y = np.array([1, 1, 1, 0, 0, 0])
```

## (a) decision boundary for 1-NN based on these data points.

```python
plot_decision_boundaries(X, y, KNeighborsClassifier, n_neighbors=1)
```

## (b) decision boundary for 3-NN based on these data points.

```python
plot_decision_boundaries(X, y, KNeighborsClassifier, n_neighbors = 3)
```

### 5. Support Vector Machine

(a) No.
SVM's motivation is to find an optimal decision boundary, so that there is a margin between the decision boundary and each group of data, and the margin on each side needs to be maximized. Therefore, the greater the distance from the decision boundary to each training example, the smaller the error rate in the classification operation.
In this case, no matter which one of the non circled point is removed, the decision boundary will not be altered.


(b) Generally speaking, hard margin has a strict constraint to correctly classify each and every data point. If our points are linearly separable then only our hyperplane is able to distinguish between them. However, if there are any outliers, it is not able to be separated.  Usually, using hard margins in our real life is not so realistic. Therefore, we need to use soft margins to skip some outliers and be able to classify almost linearly separable points, we can tolerate certain errors with the more loose conditions. 
 
(c) The gamma parameter defines how far the influence of a single training example reaches. The low value of gamma meaning is ‘far’ and high values of gamma meaning ‘close’. The higher gamma, the closer other examples are to be affected. 

The C parameter is more like a regularization parameter in the SVM. It trades off the correct classification of training examples against maximization of the decision function’s margin. If the decision function is better at classifying all training points correctly, we will have a larger C and smaller margin be accepted. Similarly, a smaller c will encourage a larger margin which will make the decision function more simple but lose some training accuracy.
